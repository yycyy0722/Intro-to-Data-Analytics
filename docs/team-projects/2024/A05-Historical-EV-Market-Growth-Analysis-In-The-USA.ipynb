**Columns**:

Example: Each row represents a EV register, each column contains register information described as below:


*   **State (12 State)**: The U.S. state where the vehicle is registered, selected from a group of 12 states.

*   **ZIP Code**: The postal code for the area where the vehicle is registered.

*   **Registration Data**: Information related to the vehicle’s official registration with the Department of Motor Vehicles (DMV).

*   **Vehicle Make**: The manufacturer or brand of the vehicle (e.g., Tesla, Ford).
*   **Vehicle Model**: The specific model of the vehicle (e.g., Model 3, F-150).

*   **Vehicle Model Year**: The year in which the vehicle model was manufactured.

*   **Drivetrain Type**: The type of vehicle drivetrain, such as front-wheel drive (FWD), rear-wheel drive (RWD), or all-wheel drive (AWD).

*   **Vehicle GVWR Class**: The Gross Vehicle Weight Rating (GVWR) class that categorizes the vehicle based on its weight.

*   **Vehicle GVWR Category**: The specific weight category the vehicle falls into, based on its GVWR class (e.g., light, medium, heavy).

*   **Vehicle Count**: The total number of vehicles counted for a specific set of data.
*   **DMV Snapshot ID**: A unique identifier for a particular set of DMV data records.
*   **DMV Snapshot (Date)**: The date when the DMV took the snapshot of the registration data.

*   **Latest DMV Snapshot Flag**: A marker indicating whether the snapshot is the most recent available.

*   **County**: The administrative division within the state where the vehicle is registered.

Alternative fueling stations play a critical role in supporting the transition to cleaner energy sources in the transportation sector. This dataset provides detailed information on the availability of different types of fueling stations across the United States, offering insights into the infrastructure that supports alternative fuels such as electricity, hydrogen, and biodiesel.

Companies and policymakers can use this data to identify regions with the strongest alternative fueling infrastructure, which helps guide future investments and expansion plans. Additionally, it supports environmental goals by providing information on the accessibility of cleaner fuel options across different states.

```python
# Code to pull from google cloud

# This code was modified from Professor Soltanieh-ha's example and subsequently modified for our example
def download_blob(project_id, bucket_name):
    """Downloads a blob from the bucket."""
    # project_id = "your-project-id"
    # bucket_name = "your-bucket-name"

    storage_client = storage.Client(project_id)

    bucket = storage_client.bucket(bucket_name)

    blobs = bucket.list_blobs()

    # Print the file names
    for blob in blobs:
        file_name = blob.name
        print(f"Downloading {file_name}...")
        blob.download_to_filename(file_name)
        print(f"{file_name} downloaded successfully.")
```

As we can see, there are clearly some alignment issues that persist throughout the file. Here are the different data issues that we encountered with this file:

**NA Appearing Instead of Blanks**
* On the original website, certain fields were left blank, but when we downloaded the dataset, these blank values were automatically filled with "NA." This created confusion and potentially misleading entries in our analysis. During the data cleaning process, we replaced all "NA" entries with blank spaces to reflect the original dataset accurately. This step ensured that missing data was handled appropriately without introducing false values, thereby preserving the integrity of our analysis.

**Misplaced Data**
* Some data entries were placed in the wrong rows or columns, leading to inconsistencies across the dataset. We conducted a thorough manual review to realign and reposition misplaced data, ensuring that each value matched the correct corresponding field.

**State Column Issue**
* The "State" column was problematic, with state names appearing in incorrect rows or being associated with the wrong data. We fixed these inconsistencies by carefully reviewing and realigning the state information, ensuring that each entry was correctly tied to the appropriate data points.

**Notes at the Bottom**
* The last four lines of the file contained notes from the original dataset, which were not useful for our analysis. We removed these lines to clean up the dataset, ensuring that only relevant data remained for further analysis.


Since each sheet has the same data cleaning that needs to be done. We utilize a for loop to go through each sheet of our file. Once each sheet has been cleaned, we concatenate them into one csv which we will use later on for the visualization and analysis component.

```python
# Change the first row to be the column headers
for sheet in station_year_df:
    station_year_df[sheet].columns = station_year_df[sheet].iloc[0]
    station_year_df[sheet] = station_year_df[sheet].rename(columns={"Electrica" : "Electric",
                                                                    "Hydrogenb": "Hydrogen",
                                                                    "Propanec": "Propane",
                                                                    "Totald" : "Total of EV Outlets and All Other Stations"})

    # Can change the drop to a sequence to grab multiple rows or cane make it a condition
    station_year_df[sheet] = station_year_df[sheet].drop([0,1,2])
    station_year_df[sheet].insert(0,"Year",sheet)
    station_year_df[sheet].insert(5,"Electric Stations","S")
    station_year_df[sheet].insert(6,"Electric Charging Outlets","CO")
    station_year_df[sheet].insert(7,"Level 1",1)
    station_year_df[sheet].insert(8,"Level 2",2)
    station_year_df[sheet].insert(9,"DC Fast","DC Fast")

    # Split the columns at Electric and paste to the three corresponding columns
    # Did this so I can cast these up a row and move on
    station_year_df[sheet][['Level 1', 'Level 2', 'DC Fast']] = station_year_df[sheet]['Electric'].str.split('|', expand=True)

    # Shift the relevant columns back to the previous row when 'State' is NaN --> Assistance from ChatGpt in this code
    mask = station_year_df[sheet]['State'].isna()
    station_year_df[sheet].loc[mask.shift(-1).fillna(False), ['Level 1', 'Level 2', 'DC Fast']] = station_year_df[sheet][['Level 1', 'Level 2', 'DC Fast']].shift(-1)

    # Drop the NA States
    station_year_df[sheet] = station_year_df[sheet].dropna(subset=['State'])

    # Redo the Electric Stations/ Electric Charging Outlets split
    station_year_df[sheet][['Electric Stations', 'Electric Charging Outlets']] = station_year_df[sheet]['Electric'].str.split('|', expand=True)

    # Drop the Electric column as it is redundant
    station_year_df[sheet] = station_year_df[sheet].drop(columns=['Electric'])

    # Drop last 4 rows and Total row because they are not needed
    station_year_df[sheet] = station_year_df[sheet].dropna()
    station_year_df[sheet] = station_year_df[sheet].drop(station_year_df[sheet].index[-1:])

    # Take the total of the Hydrogen column
    station_year_df[sheet]['Hydrogen'] = station_year_df[sheet]['Hydrogen'].str.split('|', expand=False).str.get(2)

    # Take the total of the Propane Column
    station_year_df[sheet]['Propane'] = station_year_df[sheet]['Propane'].str.split('|', expand=False).str.get(2)


    # Convert the data types and gets rid of the commas for values that are greater than 1,000
    station_year_df[sheet]['Year'] =  station_year_df[sheet]['Year'].astype(str).str.replace(',', '').str.strip().astype(int)
    station_year_df[sheet]['Biodiesel'] =  station_year_df[sheet]['Biodiesel'].astype(str).str.replace(',', '').str.strip().astype(int)
    station_year_df[sheet]['CNG'] =  station_year_df[sheet]['CNG'].astype(str).str.replace(',', '').str.strip().astype(int)
    station_year_df[sheet]['E85'] =  station_year_df[sheet]['E85'].astype(str).str.replace(',', '').str.strip().astype(int)
    station_year_df[sheet]['Electric Stations'] =  station_year_df[sheet]['Electric Stations'].astype(str).str.replace(',', '').str.strip().astype(int)
    station_year_df[sheet]['Electric Charging Outlets'] =  station_year_df[sheet]['Electric Charging Outlets'].astype(str).str.replace(',', '').str.strip().astype(int)
    station_year_df[sheet]['Level 1'] =  station_year_df[sheet]['Level 1'].astype(str).str.replace(',', '').str.strip().astype(int)
    station_year_df[sheet]['Level 2'] =  station_year_df[sheet]['Level 2'].astype(str).str.replace(',', '').str.strip().astype(int)
    station_year_df[sheet]['DC Fast'] =  station_year_df[sheet]['DC Fast'].astype(str).str.replace(',', '').str.strip().astype(int)
    station_year_df[sheet]['Renewable Diesel'] =  station_year_df[sheet]['Renewable Diesel'].astype(str).str.replace(',', '').str.strip().astype(int)
    station_year_df[sheet]['Hydrogen'] =  station_year_df[sheet]['Hydrogen'].astype(str).str.replace(',', '').str.strip().astype(int)
    station_year_df[sheet]['Propane'] =  station_year_df[sheet]['Propane'].astype(str).str.replace(',', '').str.strip().astype(int)
    station_year_df[sheet]['LNG'] =  station_year_df[sheet]['LNG'].astype(str).str.replace(',', '').str.strip().astype(int)
    station_year_df[sheet]['Total of EV Outlets and All Other Stations'] =  station_year_df[sheet]['Total of EV Outlets and All Other Stations'].astype(str).str.replace(',', '').str.strip().astype(int)

    # Add each sheet to a list so that they can be concatenated into one sheet
    station_year_list.append(station_year_df[sheet])

# Concatenate all the data into one dataframe
all_stations_data = pd.concat(station_year_list).reset_index(drop=True)

# Save as a CSV file to be exported back to the cloud
all_stations_data.to_csv('all_station_data.csv', index=False)
```

**12) How can we compare the number of EV registrations per state will the number of charging stations after previously examining the ratio of the two above?**

Based on the previous chart, we can see that the peak growth of EV charging stations occurred between 2020 and 2021, primarily driven by California. Excluding California, other states experienced their highest growth in 2022 and 2023, which aligns with vehicle registration trends. To further analyze potential growth opportunities, I created a chart comparing the growth rates of EV registrations and charging stations by state.

```python
# 1. Sort states by total electric vehicle registrations
states_sorted = combined_ev_data.groupby('State')['Vehicle Count Complete'].sum().sort_values(ascending=False).index.tolist()

# 2. Define groups, four states per group
group1 = states_sorted[:4]
group2 = states_sorted[4:8]
group3 = states_sorted[8:12]  # Adjust according to the actual number of states
groups = [group1, group2, group3]

# 3. Create subplots, setting four subplots per row
fig, axes = plt.subplots(len(groups), 4, figsize=(24, 18), sharex=False, sharey=False)

handles, labels = [], []

# If there is only one row, axes might be a 1D array and need to be converted to 2D
if len(groups) == 1:
    axes = [axes]

for row, group in enumerate(groups):
    group_data = combined_ev_data[combined_ev_data['State'].isin(group)]

    electric_stations_min = group_data['Electric Stations'].min()
    electric_stations_max = group_data['Electric Stations'].max()
    vehicle_count_min = group_data['Vehicle Count Complete'].min()
    vehicle_count_max = group_data['Vehicle Count Complete'].max()

    station_padding = (electric_stations_max - electric_stations_min) * 0.1
    vehicle_padding = (vehicle_count_max - vehicle_count_min) * 0.1

    for col, state in enumerate(group):
        ax = axes[row, col]
        state_data = group_data[group_data['State'] == state]

        ax1 = ax
        sns.lineplot(data=state_data, x='Year', y='Electric Stations', label='Electric Stations', marker='o', ax=ax1, color='tab:blue')

        # Set y-axis label conditionally
        if col == 0:
            ax1.set_ylabel('Electric Stations', fontsize=12, color='tab:blue')
            ax1.tick_params(axis='y', labelcolor='tab:blue')
        else:
            ax1.set_ylabel('')
            ax1.tick_params(axis='y', labelcolor='tab:blue', labelleft=False)

        ax2 = ax1.twinx()
        sns.lineplot(data=state_data, x='Year', y='Vehicle Count Complete', label='Vehicle Count Complete', marker='o', ax=ax2, color='tab:orange')

        # Set y-axis label conditionally
        if col == 3:
            ax2.set_ylabel('Vehicle Count Complete', fontsize=12, color='tab:orange')
            ax2.tick_params(axis='y', labelcolor='tab:orange')
        else:
            ax2.set_ylabel('')
            ax2.tick_params(axis='y', labelcolor='tab:orange', labelright=False)

        ax1.set_ylim(electric_stations_min - station_padding, electric_stations_max + station_padding)
        ax2.set_ylim(vehicle_count_min - vehicle_padding, vehicle_count_max + vehicle_padding)

        ax1.set_title(state, fontsize=14)
        ax1.set_xlabel('Year', fontsize=12)
        ax1.grid(True)

        if row == 0 and col == 3:
            for line in ax1.get_lines():
                handles.append(line)
                labels.append(line.get_label())
            for line in ax2.get_lines():
                handles.append(line)
                labels.append(line.get_label())

            # Create a single legend outside of the subplots
            fig.legend(handles, labels, loc='upper right', fontsize=12, title='Legend', bbox_to_anchor=(1, 1), borderaxespad=0.)

        else:
            ax1.legend_.remove()
            ax2.legend_.remove()

# If the total number of states is not a multiple of the group size, remove extra subplots
total_states = len(groups) * 4
current_states = len(states_sorted)
if current_states < total_states:
    for i in range(current_states, total_states):
        fig.delaxes(axes.flatten()[i])

# Add a main title
plt.tight_layout(rect=[0, 0, 0.95, 0.96])
fig.suptitle('Electric Stations and Electric Vehicles Registered by State', fontsize=20)

# Display the plot
plt.show()
```

**14) Is the growth rate of EV stations adequate to support the growing EV cars? What is the density of EV stations? How many cars does 1 station have to cater?**

This is an important metric because EV cars take 5 time more time to charge/fuel up than regular gasoline/ diesel cars.

```python
# Make this copy of all_station_data so that we do not mess with the original dataset
usa_states = all_station_data.copy()
all_state_data=usa_states.merge(lv_regi_state_data, on = ['State','Year'], how='left').sort_values(by=['Year','State'])

# Calculate the density of EV
all_state_data['EV_station_density']=all_state_data['Electric (EV)']/all_state_data['Electric Stations']

# Define a Custom Green-to-Red Color Scale
custom_color_scale = [
    (0.0, 'green'),
    (0.5, 'yellow'),
    (1.0, 'red')
]

# Determine US zmin and zmax for consistent color scaling
us_zmin = all_state_data['EV_station_density'].min()
us_zmax = all_state_data['EV_station_density'].max()

# Create the Choropleth Map of Car Density
fig = px.choropleth(
    all_state_data,
    locations='StateAbr',
    locationmode='USA-states',
    color='EV_station_density',
    scope='usa',
    color_continuous_scale=custom_color_scale,
    range_color=(us_zmin, us_zmax),
    animation_frame='Year',
    animation_group='StateAbr',
    hover_name='State',
    hover_data={'StateAbr': False, 'Year': True, 'EV_station_density': True},
    labels={'Density': 'Cars per Station'},
    title='EV Cars Density per Charging Station by State'
)

fig.show()
```
