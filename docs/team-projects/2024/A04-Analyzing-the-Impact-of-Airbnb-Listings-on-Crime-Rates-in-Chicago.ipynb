## 3. Data Overview

### 3.1 Overview
This data came from two different data sources. The first one is Airbnb listing data from 'Inside Airbnb' website: https://insideairbnb.com/. In the Airbnb data, it provides information on each property listed on Airbnb in Chicago. Some key attributes usually include:

1. Property Details: Information such as the type of property (e.g., apartment, house), room type (entire home or shared room), number of bedrooms, bathrooms, and other amenities.

2. Host Information: Data on the host, including their location, response rate, whether they are designated as a superhost, and their experience on the platform.

3. Pricing and Availability: Nightly price of the property, additional fees (e.g., cleaning fee), availability for future bookings, and discounts for extended stays.

5. Location: Geographical information such as latitude, longitude, and neighborhood, which is key for spatial analysis.

The second dataset came from the Chicago Data Portal: https://data.cityofchicago.org/. The Chicago crime dataset typically includes detailed records of reported crimes in the city of Chicago. Some key attributes include:

1. Crime Type and Description: General classification of the crime (e.g., theft, assault) and a more detailed description of the incident.

2. Location: The specific street block where the crime occurred, as well as the latitude and longitude for geographic analysis. Other geographic identifiers 
include the community area, ward, and police district.

3. Date and Time: Information about when the crime occurred, which allows for temporal analysis to identify trends by year, month, or time of day.

4. Arrest and Resolution: Data on whether an arrest was made and other relevant details about how the crime was resolved.

5. Domestic Incident: Indicator of whether the crime was related to domestic violence, which can be a significant factor in some areas.

```python
# List of features to be used in the model (include relevant numeric and categorical features)
features = ['bedrooms', 'bathrooms', 'beds', 'room_type', 'neighbourhood_cleansed']

# Drop rows where any of the selected features are missing, as we cannot use them for prediction
airbnb_data_clean = airbnb_data.dropna(subset=features)

# Separate rows with and without missing 'price'
not_missing = airbnb_data_clean.dropna(subset=['price'])
missing = airbnb_data_clean[airbnb_data_clean['price'].isnull()]

# One-hot encode categorical variables for training (e.g., 'room_type' and 'neighbourhood_cleansed')
X_train = pd.get_dummies(not_missing[features], drop_first=True)

# Target variable for training
y_train = not_missing['price']

# One-hot encode the categorical variables in the missing data
X_missing = pd.get_dummies(missing[features], drop_first=True)

# Reindex the missing data to ensure it has the same columns as the training data
X_missing = X_missing.reindex(columns=X_train.columns, fill_value=0)
```

```python
# Initialize the regression model
model = LinearRegression()

# Train the model on the available price data
model.fit(X_train, y_train)

# Use the trained model to predict the missing prices
missing['price'] = model.predict(X_missing)

# Fill the missing prices in the original dataset
airbnb_data.loc[airbnb_data['price'].isnull(), 'price'] = missing['price']
```

### 5.3 Summary

In the data cleaning phase, we first addressed the missing longitude and latitude values in the Chicago Crime data. We filled part of the missing values by using the median longitude and latitude, grouped by block, since multiple crime incidents occurring on the same block often lacked coordinates. This method efficiently filled a portion of the missing data. For the remaining missing longitude and latitude values, we used the Geocoding API to populate their coordinates based on their block. For other missing values, if they represented a relatively small number of rows compared to the entire dataset or involved duplicate columns, we chose to drop them.

Since Location is just a combination of Longitude and Latitude, and X and Y Coordinates are two different ways of specifying locations on the Earth or a map (Latitude/Longitude - Geographic Coordinate System (GCS), X/Y Coordinates - Projected Coordinate System (PCS)), we have decided to drop these columns before merging.

For the Airbnb data, we initially focused on missing rating values. We filled these missing ratings by calculating the mean rating for each neighborhood. Several columns, such as the first review date, last review date, and property descriptions, were also blank, likely because no reviews were available or the information wasn't provided. Columns like host response rate or the number of bathrooms and bedrooms were similarly incomplete, so we filled these missing values with "Unknown" or "Not Provided." One special imputation we performed was predicting the missing property prices using simple linear regression, based on features such as bedrooms, bathrooms, beds, room_type, and neighbourhood_cleansed.

```python
# delet useless columns in aiebnb_data
airbnb_data_select = airbnb_data.drop(columns=['last_scraped', 'name', 'source', 'description', 'host_since', 'neighborhood_overview', 'host_name', 'host_about',
                                              'listing_url', 'picture_url', 'host_url', 'host_thumbnail_url', 'host_picture_url', 'host_neighbourhood',
                                               'neighbourhood', 'host_verifications', 'host_has_profile_pic', 'host_identity_verified',
                                               'minimum_minimum_nights', 'maximum_minimum_nights', 'minimum_maximum_nights', 'maximum_maximum_nights',
                                               'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm', 'has_availability',
                                               'calendar_last_scraped', 'license', 'instant_bookable', 'amenities', 'calculated_host_listings_count', 
                                               'calculated_host_listings_count_entire_homes', 'calculated_host_listings_count_private_rooms',
                                               'calculated_host_listings_count_shared_rooms'])
# format'date_column'
crime_data['Date'] = pd.to_datetime(crime_data['Date'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')

# select case happen after year 2022
crime_data = crime_data[crime_data['Year'] > 2023]

crime_data_select = crime_data.drop(columns=['IUCR', 'Description', 'FBI Code', 'Updated On', 'Location Description', 'Beat', 'District',
                   'Ward', 'Community Area'])
crime_data_select = crime_data_select[crime_data_select['Date'] <= '2024-06-21']
```

```python
# Haversine formula (vectorized version)
def haversine_vectorized(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    dlon = lon2[:, np.newaxis] - lon1
    dlat = lat2[:, np.newaxis] - lat1

    a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2[:, np.newaxis]) * np.sin(dlon / 2) ** 2
    c = 2 * np.arcsin(np.sqrt(a))

    r = 3956  # Earth's radius (miles)
    return c * r

# Optimized function to find nearby locations
def find_nearby_optimized(df1, df2, radius):
    lat1 = df1['latitude'].values
    lon1 = df1['longitude'].values
    lat2 = df2['Latitude'].values
    lon2 = df2['Longitude'].values

    # Precompute the bounding box for all df1 entries
    lat_min = lat1 - (radius / 69)
    lat_max = lat1 + (radius / 69)
    lon_min = lon1 - (radius / (69 * np.cos(np.radians(lat1))))
    lon_max = lon1 + (radius / (69 * np.cos(np.radians(lat1))))

    results = []

    for i in range(len(df1)):
        # Create a mask for df2 to filter by the bounding box
        mask = (df2['Latitude'] >= lat_min[i]) & (df2['Latitude'] <= lat_max[i]) & \
               (df2['Longitude'] >= lon_min[i]) & (df2['Longitude'] <= lon_max[i])
        filtered_df2 = df2[mask]

        if not filtered_df2.empty:
            # Compute distances only for filtered points
            distances = haversine_vectorized(lon1[i], lat1[i], filtered_df2['Longitude'].values, filtered_df2['Latitude'].values)
            nearby_indices = np.where(distances <= radius)[0]

            if nearby_indices.size > 0:
                matched_df1 = pd.DataFrame([df1.iloc[i]] * len(nearby_indices), columns=df1.columns)
                matched_df2 = filtered_df2.iloc[nearby_indices].reset_index(drop=True)
                results.append(pd.concat([matched_df1.reset_index(drop=True), matched_df2], axis=1))

    # Concatenate results into a single DataFrame
    return pd.concat(results, ignore_index=True) if results else pd.DataFrame()  # Return empty DataFrame if no matches found

merged_df = find_nearby_optimized(airbnb_data_select, crime_data_select, radius=0.5)
```

### 6.2 Summary

When dealing with datasets that lack common columns for direct merging, an effective strategy is to calculate distances using the Haversine formula, which is based on geographic coordinates (latitude and longitude). This approach allows for the identification of nearby data points, enabling the grouping of these points into the same neighborhood. By using this method, we can effectively merge the datasets despite the absence of shared attributes, ensuring that related information is combined in a meaningful way. This technique not only facilitates data integration but also enhances the overall analysis by providing context based on geographical proximity.

### 7.11 How crime ratio affect guests satisfaction ?

```python
# Step 1: Define the threshold for high-rating Airbnb listings
rating_threshold = 4.5

# Create a new column to classify listings as high-rating or low-rating
merge_data['rating_group'] = merge_data['review_scores_location'].apply(lambda x: 'High' if x >= rating_threshold else 'Low')

# Step 2: Define the list of crime types you are interested in analyzing
crime_types_of_interest = ['CRIMINAL TRESPASS', 'BURGLARY', 'BATTERY', 'THEFT', 'WEAPONS VIOLATION', 'ROBBERY']

# Step 3: Filter the data to only include the crimes of interest
crime_data_of_interest = merge_data[merge_data['Primary Type'].isin(crime_types_of_interest)]

# Step 4: Group by rating group and crime type to calculate the crime proportion for each rating group
crime_ratio_by_rating = crime_data_of_interest.groupby(['rating_group', 'Primary Type']).size().reset_index(name='crime_count')

# Step 5: Calculate the total number of listings in each rating group
total_listings_by_rating = merge_data.groupby('rating_group').size().reset_index(name='total_listings')

# Step 6: Merge the total listings with the crime count by rating group
crime_ratio_by_rating = pd.merge(crime_ratio_by_rating, total_listings_by_rating, on='rating_group', how='left')

# Step 7: Calculate the crime ratio for each crime type within the rating group
crime_ratio_by_rating['crime_ratio'] = (crime_ratio_by_rating['crime_count'] / crime_ratio_by_rating['total_listings'])

# Pivot the data for easier visualization
crime_ratio_pivot = crime_ratio_by_rating.pivot(index='Primary Type', columns='rating_group', values='crime_ratio')

# order the display
crime_ratio_pivot = crime_ratio_pivot.sort_values(by='High', ascending=False)

# Plot the crime ratios for high vs low-rated areas
ax = crime_ratio_pivot.plot(kind='bar', figsize=(10, 6))

plt.title('Crime Ratios in High-Rated vs Low-Rated Airbnb Listings')
plt.xlabel('Primary Crime Type')
plt.ylabel('Crime Ratio')
plt.legend(title='Rating Group')
plt.grid(True)
plt.show()
```

#### Answer : 
Violent crimes such as battery and weapons violations have a strong negative impact on guest satisfaction, contributing to lower Airbnb ratings.
Theft is more prevalent in high-rated listings, indicating that guests might tolerate theft if the area offers other desirable factors such as proximity to popular attractions or vibrant environments.
Robbery has a moderate negative effect on ratings, while trespass and burglary seem to have a smaller impact on guest satisfaction.
